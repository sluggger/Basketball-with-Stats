```python
stats['PAVG'] = stats['PTS']/stats['Games Played']
stats['PAVG']
```

```python
categorical_columns = ['Name', 'Games Played', 'MIN', 'PTS', 'FGM', 'FGA', 'FG%', '3PM', '3PA',
       '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK',
       'TOV', 'PF', 'EFF', 'AST/TOV', 'STL/TOV', 'Age', 'Birth_Place',
       'Birthdate', 'Collage', 'Experience', 'Height', 'Pos', 'Team', 'Weight',
       'BMI']
for column in categorical_columns:
    stats[column] = stats[column].astype('category')
    
stats[categorical_columns].describe()
```

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.models import HoverTool
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
from bokeh.models import HoverTool

x = stats["MIN"]
y= stats["PTS"]
Name = stats["Name"]

fig = figure(
    tools="pan,box_zoom,wheel_zoom,zoom_in,zoom_out,reset,save",
    title="Example Bokeh plot",
    y_range=[0, 2000],
    x_axis_label='Minutes',
    y_axis_label='Points',
    width=600, height=400,
)

scatter = fig.circle(stats["MIN"],stats["PTS"], legend_label="y=x", fill_color="white", size=4)
tooltips = [('Value', '@Name')]
hover_tool = HoverTool(renderers=[scatter], tooltips=tooltips)
fig.add_tools(hover_tool)

show(fig)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

x = stats["MIN"]
y = stats["PTS"]
name = stats["Name"]

source = ColumnDataSource(data=dict(x=x, y=y, name=name))

fig = figure(
    tools="pan, box_zoom, wheel_zoom, zoom_in, zoom_out, reset, save",
    title="Example Bokeh plot",
    y_range=[0, 2000],
    x_axis_label='Minutes',
    y_axis_label='Points',
    width=600,
    height=400,
)

scatter = fig.circle('x', 'y', source=source, legend_label="y=x", fill_color="white", size=4)
hover_tool = HoverTool(renderers=[scatter], tooltips=[('Name', '@name')])
fig.add_tools(hover_tool)

show(fig)
```

```python
x = stats["Age"]
y= stats["MIN"]

fig = figure(
    tools="pan,box_zoom,wheel_zoom,zoom_in,zoom_out,reset,save",
    title="Example Bokeh plot",
    y_range=[0, 2000],
    x_axis_label='Age',
    y_axis_label='Minutes',
    width=600, height=400,
)

scatter = fig.circle(x, y, legend_label="y=x", fill_color="white", size=4)
tooltips = [('Value', '@MIN')]
hover_tool = HoverTool(renderers=[scatter], tooltips=tooltips)
fig.add_tools(hover_tool)


show(fig)
```

```python
import plotly.graph_objects as go

x = stats["3PA"]
y = stats["3PM"]
name = stats["Name"]  # Additional value to show on hover

fig = go.Figure(
    layout=dict(
        title="Shooters gonna shoot",
        yaxis_range=[0, 300],
        xaxis=dict(title='3PM'),
        yaxis=dict(title='3PA'),
    ),
)

# Add hovertemplate to show 'Name' value on hover
scatter = go.Scatter(x=x, y=y, mode='markers', name="y=x", marker=dict(color='purple', size=8),
                     hovertemplate='Name: %{text}<br>3PA: %{x}<br>3PM: %{y}')
fig.add_trace(scatter)

fig.update_traces(text=name)  # Set 'Name' as the hover text

fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({'x': [1, 2, 3, 4, 5],
                     'y': [6, 7, 2, 4, 5],
                     'valueses': ['A', 'B', 'C', 'D', 'E']})

# Create a figure
p = figure()

# Plot the scatter plot
scatter = p.scatter(x='x', y='y', size=10, color='red', source=data)

# Add tooltips
tooltips = [('Valueeee', '@valueses')]
hover_tool = HoverTool(renderers=[scatter], tooltips=tooltips)
p.add_tools(hover_tool)

# Show the graph
show(p)
```
